```python
# caption_single.py
from PIL import Image, ImageOps
import torch
from transformers import VisionEncoderDecoderModel, ViTImageProcessor, AutoTokenizer
from IPython.display import display

# ----- Config -----
MODEL_NAME = "nlpconnect/vit-gpt2-image-captioning"
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
MAX_LENGTH = 64
NUM_BEAMS = 4
DO_SAMPLE = False   # True enables sampling / diversity
TOP_K = None
TOP_P = None

# ----- Load model -----
model = VisionEncoderDecoderModel.from_pretrained(MODEL_NAME)
processor = ViTImageProcessor.from_pretrained(MODEL_NAME)
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model.to(DEVICE)
model.eval()

# ----- Caption function -----
def generate_caption_from_pil(image_pil):
    # preprocess
    if image_pil.mode != "RGB":
        image_pil = image_pil.convert("RGB")
    # optional: resize for speed (model processor will handle)
    pixel_values = processor(images=image_pil, return_tensors="pt").pixel_values.to(DEVICE)

    gen_kwargs = {
        "max_length": MAX_LENGTH,
        "num_beams": NUM_BEAMS,
        "do_sample": DO_SAMPLE
    }
    if DO_SAMPLE:
        if TOP_K: gen_kwargs["top_k"] = TOP_K
        if TOP_P: gen_kwargs["top_p"] = TOP_P

    output_ids = model.generate(pixel_values, **gen_kwargs)
    caption = tokenizer.decode(output_ids[0], skip_special_tokens=True).strip()
    return caption

# ----- Example: load image locally (replace path) -----
img_path = "example.jpg"   # replace with your image filename
img = Image.open(img_path)
display(img)
print("Caption:", generate_caption_from_pil(img))
```

```python
# Colab cell
from PIL import Image
import torch
from transformers import VisionEncoderDecoderModel, ViTImageProcessor, AutoTokenizer
from IPython.display import display, HTML
from google.colab import files
from pathlib import Path
import io
import sys

# CONFIG - change if you want a different model
MODEL_NAME = "nlpconnect/vit-gpt2-image-captioning"  # beginner-friendly
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print("Device:", DEVICE)

# Load model, processor, tokenizer
# This downloads model weights the first time you run it
model = VisionEncoderDecoderModel.from_pretrained(MODEL_NAME)
processor = ViTImageProcessor.from_pretrained(MODEL_NAME)
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model.to(DEVICE)
model.eval()

print("Model loaded:", MODEL_NAME)
```

```python
# Colab cell
# Generation settings you (or your interviewer) can tweak
MAX_LENGTH = 64
NUM_BEAMS = 4        # set to 1 for fastest but less accurate; 4-6 is common
DO_SAMPLE = False    # False => deterministic beam search. True => sampling for diversity.
TOP_K = 50           # used if DO_SAMPLE is True
TOP_P = 0.9          # nucleus sampling, used if DO_SAMPLE is True
NUM_RETURN_SEQUENCES = 1  # how many captions to produce per image

def generate_captions_from_pil(image_pil,
                               max_length=MAX_LENGTH,
                               num_beams=NUM_BEAMS,
                               do_sample=DO_SAMPLE,
                               top_k=TOP_K,
                               top_p=TOP_P,
                               num_return_sequences=NUM_RETURN_SEQUENCES):
    """
    Returns a list of captions (strings).
    """
    if image_pil.mode != "RGB":
        image_pil = image_pil.convert("RGB")
    # Preprocess image(s)
    pixel_values = processor(images=image_pil, return_tensors="pt").pixel_values.to(DEVICE)
    gen_kwargs = {
        "max_length": max_length,
        "num_beams": num_beams if not do_sample else 1,  # beams don't mix with sampling well
        "do_sample": do_sample,
        "num_return_sequences": num_return_sequences
    }
    if do_sample:
        gen_kwargs["top_k"] = top_k
        gen_kwargs["top_p"] = top_p

    # generate
    with torch.no_grad():
        output_ids = model.generate(pixel_values, **gen_kwargs)

    # if multiple sequences returned, output_ids shape reflects that
    captions = [tokenizer.decode(g, skip_special_tokens=True).strip() for g in output_ids]
    return captions
```

```python
# Colab cell
print("Upload one or more image files in the dialog that appears.")
uploaded = files.upload()  # opens a file picker

if not uploaded:
    print("No files uploaded.")
else:
    out_lines = []
    for filename, filebytes in uploaded.items():
        try:
            # Load image from bytes (so it works with file uploader)
            img = Image.open(io.BytesIO(filebytes))
            display(img)  # show the image for confirmation

            # Generate captions
            captions = generate_captions_from_pil(img)
            # Print nicely
            print(f"\nFile: {filename}")
            for i, c in enumerate(captions, start=1):
                print(f" Caption {i}: {c}")

            # Save to text results
            for i, c in enumerate(captions, start=1):
                out_lines.append(f"{filename}\tcaption_{i}\t{c}\n")

        except Exception as e:
            print(f"Failed processing {filename}: {e}", file=sys.stderr)

    # Optionally save captions to a file for download
    if out_lines:
        result_path = "captions_results.txt"
        Path(result_path).write_text("".join(out_lines), encoding="utf-8")
        print(f"\nSaved captions to {result_path} — you can download it.")
        files.download(result_path)
```
